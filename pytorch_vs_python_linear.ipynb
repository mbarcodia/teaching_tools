```python
# Python Numpy

import numpy as np

# Initialize data
X = np.array([[1.0, 2.0], [3.0, 4.0]])
W = np.array([[0.5], [0.3]])
y_true = np.array([[2.0], [4.0]])

# Forward pass
y_pred = X @ W  # Matrix multiplication

# Compute loss (Mean Squared Error)
loss = np.mean((y_pred - y_true) ** 2)

# Backward pass - MANUAL gradient calculation
grad_loss = 2 * (y_pred - y_true) / y_true.shape[0]
grad_W = X.T @ grad_loss  # Chain rule by hand

# Update weights
learning_rate = 0.01
W = W - learning_rate * grad_W

print(f"Loss: {loss}")
print(f"Gradients: {grad_W}")
```

```python
# PyTorch Autograd

import torch

# Initialize data
X = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
W = torch.tensor([[0.5], [0.3]], requires_grad=True)
y_true = torch.tensor([[2.0], [4.0]])

# Forward pass
y_pred = X @ W  # Matrix multiplication

# Compute loss
loss = torch.mean((y_pred - y_true) ** 2)

# Backward pass - AUTOMATIC gradient calculation
loss.backward()

# Update weights
learning_rate = 0.01
W.data = W.data - learning_rate * W.grad

print(f"Loss: {loss.item()}")
print(f"Gradients: {W.grad}")
```
